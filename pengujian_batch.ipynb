```python
import csv
from ipynb.fs.full.tokenizing import tokenizing
from ipynb.fs.full.naivebayes import naivebayes

true=0
count=0

from ipynb.fs.full.tokenizing import tokenizing
from ipynb.fs.full.naivebayes import naivebayes

with open("datatest/datatest.csv","r",encoding='utf8') as csv_file:    
    csv_read = csv.reader(csv_file,delimiter=',')
    for line in csv_read:
        count+=1
        prediksi = naivebayes(tokenizing(line[1]))
        if prediksi == line[0]:
            true+=1
        print("No. "+str(count)+" Prediksi :"+str(prediksi)+" Jawaban : "+str(line[0]))
    print("Jawaban Benar : "+ str(true))
    print("Akurasi Prediksi: "+str(true/count*100)+"%")
```
