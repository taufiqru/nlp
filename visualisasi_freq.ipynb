```python
import pandas as pd
import mysql.connector

conn = mysql.connector.connect(host='localhost',user='root',password='',database='nlp2')
cursor = conn.cursor()
query="select * from freq_token_misc limit 10"
cursor.execute(query)
result = cursor.fetchall()
print(result)

freq_tokens = pd.DataFrame(result)
freq_tokens.columns = ['Token','Frequency'] 
freq_tokens.plot(x="Token",y="Frequency",kind='bar',title="Top 10 Token Freq on Misc")

conn.close()
```
